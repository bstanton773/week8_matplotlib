### Data Visualization <br>
<p>Data visualization is one of the most powerful tools an analyst has for two main reasons. Firstly, it is unrivalled in it's ability to guide the analysts hand in determining 'what to look at next'. Often a visual is revealing of patterns in the data that are not easily discernable by just looking at dataframes. Secondly, they are an analysts greatest communication tool. Very often professional analysts need to present their results to groups of people responsible for taking action based on what the data says. Visuals can tell your story much better then raw numbers, so let's talk a little bit about how to make effective visuals that make a point.</p>

### Different Types of Charts <br>
<ul>
    <li><b>Line Chart:</b> Exploring data over time</li>
    <li><b>Bar Chart:</b> Comparing categories of data and tracks changes over time</li>
    <li><b>Pie Chart:</b> Explore parts of a whole, i.e. fractions</li>
    <li><b>Scatter Plot:</b> Similar to line charts, track correlations between two categories</li>
    <li><b>Histogram:</b> Unrelated from bar charts, show distribution of variables</li>
    <li><b>Candlestick Chart:</b> Used a lot in financial sector, i.e. can compare a stock over a period of time</li>
    <li><b>Box Chart:</b> Looks identical to candlestick charts, compares minimum, 1st, medium, 3rd quartiles, and max values</li>
</ul>

<p>Depending on what you need to accomplish in conceptualizing your data, you will go with a specific type of chart to portray your data.</p>

### Subplots <br>
<p>Using the subplots method, you're able to create multiple plots within a single cell.</p>

```python
# first plot
plot_1 = plt.subplot(2, 1, 1) # 2 rows, 1 column, this plot is the index-1 position aka 1st position
plt.plot([1600,1700,1800,1900,2000], [.8, .9, 1.1, 1.0, 1.4], 'gx-.', label='First Demo')

plt.legend()
plt.xlabel('Year')
plt.ylabel('Population (Millions)')
plot_1.set_xlim([1550, 2050]) # [lower bound, upper bound]
plot_1.set_ylim([0.5,2])

# second plot
plot_2 = plt.subplot(2, 1, 2) # 2 rows, 1 column, this plot is display in the 2-index position aka below the 1st
plt.plot([1600,1700,1800,1900,2000], [1.1, 1.3, 1, 1.5, 1.6], 'ko:', label='Second Demo')


plt.legend()
plt.xlabel('Year')
plt.ylabel('Population (Millions)')
plot_2.set_xlim([1550, 2050])
plot_2.set_ylim([0.5,2])


plt.show()
```

### Presentation Worthiness
<p>Before we go into how to use this as a tool for data exploration, let's talk quickly about presentation-worthiness. In the next cell we're going to introduce customization of lines, colors, ticks, titles, fonts, and grids.</p>

```python
fig, ax = plt.subplots(figsize = (20, 10))

ax.plot(x, y,
        lw=5, # Line Width
        ls='--', # Line Style
        c='r', # Color
        marker='*',
        ms=15, # Marker Size
        mfc='b', # Marker Face Color
        label='Data'
       )
plt.xlabel('X Value', fontsize=32, fontname='Times New Roman')
plt.ylabel('Y Value', fontsize=32, fontname='Times New Roman')

plt.xticks([x for x in range(0,21,2)],fontsize=28)
plt.yticks([0, 20, 40, 60, 80, 100], fontsize=28)

plt.legend(fontsize=28, loc='best')

plt.grid(True)

plt.title('Visual of F(x) Equation', fontname='Times New Roman', fontsize=40, y=1.05)

plt.show()
```

<p>It's sometimes nice to obsess a little bit over how your plots look. It's good practice paying attention to detail and it has huge payoff when you end up presenting your findings to other analysts and business parteners. Here's a more complex example of using the matplotlib api:</p>

```python
markers = ['o', 'v', '^', '*', 's']
styles = ['-', '--', '-.', ':', '-']
colors = ['#588C7E', '#F2E394', '#F2AE72', '#D96459', '#8C4646']
```

```python
x = np.linspace(0, 10, 51)

fig, ax = plt.subplots(figsize=(20,10))

# Loop over a range of 5 to create 5 different lines in our graph each being x ** i
for i in range(5):
    plt.plot(x,x**i,
             lw=5,
             ls=styles[i],
             c=colors[i],
             marker=markers[i],
             ms=15,
             label=f'x^{i}'
            )

plt.xlim([0,10])
plt.ylim([0,100])
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('X Value', fontsize=28, fontname='Gabriola')
plt.ylabel('Y Value', fontsize=28, fontname='Gabriola')

plt.legend(fontsize=20)
plt.title('y = f(x)^i', fontname='Gabriola', fontsize=36, y=1.05)

plt.grid(True)
plt.show()
```

### Scatter Plot

```python
import csv
from  itertools import groupby




colors = {
    'Iris-setosa': '#2B5B84', # Dark Blue/Purplish
    'Iris-versicolor': 'g',
    'Iris-virginica': 'purple'
}
```

#### Read in the csv

```python
def read_csv(filename):
    with open(filename, 'r') as f:
        data = list(csv.reader(f))
    return data

flowers = read_csv('iris.csv')
flowers
```

#### Plot the data

```python
fig, ax = plt.subplots(figsize=(20,10))

for species, group in groupby(flowers, lambda x: x[4]):
    list_group = list(group)
    sepal_length = [float(flower[0]) for flower in list_group]
    sepal_width = [float(flower[1]) for flower in list_group]
    plt.scatter(sepal_length, sepal_width, s=10, c=colors[species], label=species)

plt.title('Iris Dataset: Width x Length of Sepal', fontname='Times New Roman', fontsize=24, y=1.05)
plt.xlabel('Sepal Length (cm)', fontname='Times New Roman', fontsize=16)
plt.ylabel('Sepal Width (cm)', fontname='Times New Roman', fontsize=16)
plt.legend()

# Save figure to local directory
fig.savefig('iris_scatter.pdf')

plt.show()
```

### Box Plot

```python
plt.figure(figsize=(15,10))

petal_lengths = []
for species, group in groupby(flowers, lambda x: x[4]):
    petal_lengths.append([float(flower[2]) for flower in group])

plt.boxplot(petal_lengths)
plt.title('Box Plot of Iris Flowers', fontname='Arial', fontsize=24, y=1.05)
plt.axis([0,4,0,10]) #[xmin, xmax, ymin, ymax]
plt.xticks([1,2,3], ['Setosa', 'Versicolor', 'Virginica'])
plt.xlabel('Iris Species', fontname='Arial', fontsize=16)
plt.ylabel('Petal Length (cm)', fontname='Arial', fontsize=16)
plt.grid(True)

fig = plt.figure()
fig.savefig('iris_box_plot.png')
plt.show()
```

### Histograms <br>
<p>While line plots are great for visualizing trends in time series data, histograms are the king of visualizing distributions. Often, the distribution of a variable is what you're interested in, and a visualization provides a lot more information than a group of summary statistics.</p>

```python
fig, ax = plt.subplots(figsize=(16, 9)) #Subplots makes a call in the background to create a plot area

# generate random data
data = np.random.uniform(0,10,100).round(2)

# Plot a histogram
plt.hist(data, bins=10, label='Data', color='r', alpha=.75)
plt.xlabel('Value', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.title('Distribution of Random Numbers between 1-10', fontsize=32)
plt.legend(fontsize=24)
plt.show()
```

<p>Let's take a look at a mystery/randomly generated DataFrame, and plot the results</p>

```python
import pandas as pd
import numpy as np

def make_mystery_dataframe():
    customer_id = np.random.randint(1000,1200,10000)
    states = {1: 'California', 0: 'Colorado'}
    statelist = []
    agelist = []
    for i in range(10000):
        ind = np.random.randint(0,2)
        if ind == 0:
            statelist.append(states[ind])
            agelist.append(round(np.random.normal(loc = 30, scale = 3)))
        elif ind == 1:
            statelist.append(states[ind])
            agelist.append(round(np.random.normal(loc = 40, scale = 5)))
    mystery_df = pd.DataFrame.from_dict({
        'customer_id': customer_id,
        'age':agelist,
        'state': statelist
    })
    
    return mystery_df


df = make_mystery_dataframe()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# First subplot - California Histogram
ax1.hist(df[df['state'] == 'California']['age'], bins=20, label='Ages in California', color='g')
ax1.set_xlabel('Value', fontsize=24)
ax1.set_ylabel('Frequency', fontsize=24)
ax1.axis([20,60,0,1300])
ax1.legend()


# Second subplot - Colorado Histogram
ax2.hist(df[df['state'] == 'Colorado']['age'], bins=20, label='Ages in Colorado', color='b')
ax2.set_xlabel('Value', fontsize=24)
ax2.set_ylabel('Frequency', fontsize=24)
ax2.axis([20,60,0,1300])
ax2.legend()


plt.show()
```
